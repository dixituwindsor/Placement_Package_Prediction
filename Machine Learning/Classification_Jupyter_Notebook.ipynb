```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
df = pd.read_csv('Dataset.csv')
df.head()
```

```python
df_copy = df.copy()
df_copy.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_copy['gender'] = le.fit_transform(df_copy['gender'])
df_copy['ssc_b'] = le.fit_transform(df_copy['ssc_b'])
df_copy['hsc_b'] = le.fit_transform(df_copy['hsc_b'])
df_copy['workex'] = le.fit_transform(df_copy['workex'])
df_copy['gender'] = le.fit_transform(df_copy['gender'])
df_copy['specialisation'] = le.fit_transform(df_copy['specialisation'])
df_copy['status'] = le.fit_transform(df_copy['status'])
```

```python
df_copy.head()
```

```python
df_copy = df_copy.drop("sl_no", axis=1)
df_copy = df_copy.drop("salary", axis=1)
df_copy = pd.get_dummies(df_copy, columns = ['degree_t', 'hsc_s'])
df_copy.head()
```

```python
status = df_copy['status']
df_copy.drop(['status'], axis=1, inplace=True)
```

```python
gender = df_copy['gender']
ssc_b = df_copy['ssc_b']
hsc_b = df_copy['hsc_b']
workex = df_copy['workex']
specialisation = df_copy['specialisation']
df_copy.insert(16, 'status', status)
df_copy.insert(10, 'gender_encoded', gender)
df_copy.insert(11, 'ssc_b_encoded', ssc_b)
df_copy.insert(12, 'hsc_b_encoded', hsc_b)
df_copy.insert(13, 'workex_encoded', workex)
df_copy.insert(14, 'specialisation_encoded', specialisation)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
acc = []
for i in range(1,20):
    for j in range(1,150):
        rfc=RandomForestClassifier(n_estimators=i, random_state=j)
        rfc.fit(x_tr,y_tr)
        y_pre = rfc.predict(x_te)
        a = accuracy_score(y_te,y_pre)
        acc.append(a)
```

```python
max = 0
count = 1
for i in acc:
    if i > max:
        max = i
        print(i)
        print(count)
    count = count + 1    
print(count)
print('max =',max)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rfc=RandomForestClassifier(n_estimators=5, random_state=135)
rfc.fit(x_tr,y_tr)
y_pr = rfc.predict(x_te)
a = accuracy_score(y_te,y_pr)
print(a)
df = pd.DataFrame({'Actual':y_te, 'Predicted':y_pr})
df
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(x_tr,y_tr)
y_pre = svc.predict(x_te)
print(metrics.accuracy_score(y_te,y_pre))
df = pd.DataFrame({'Actual':y_te, 'Predicted':y_pre})
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
acc3 = []
for i in range(3,31):
    for j in range(1,3):
        knc= KNeighborsClassifier(n_neighbors=i, metric='minkowski', p=j)
        knc.fit(x_tr,y_tr)
        y_p = knc.predict(x_te)
        c = metrics.accuracy_score(y_te,y_p)
        acc3.append(c)
# df = pd.DataFrame({'Actual':y_te, 'Predicted':y_p})
# df
```

```python
max = 0
count = 1
for i in acc3:
    if i > max:
        max = i
        print(i)
        print(count)
    count = count + 1    
print(count)
print('max =',max)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_tr,y_tr)
y_predict = lr.predict(x_te)
print(metrics.accuracy_score(y_te,y_predict))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_tr,y_tr)
y_predi = gnb.predict(x_te)
print(metrics.accuracy_score(y_te,y_predi))
```

```python
from sklearn.neural_network import MLPClassifier
acc4 = []
for i in range(1,21):
    for j in range(1,21):
        mlpc = MLPClassifier(random_state=i, max_iter=500, hidden_layer_sizes=(j,))
        mlpc.fit(x_tr,y_tr)
        y_predic = mlpc.predict(x_te)
        d = metrics.accuracy_score(y_te,y_predic)
        acc4.append(d)
```

```python
max = 0
count = 1
for i in acc4:
    if i > max:
        max = i
        print(i)
        print(count)
    count = count + 1    
print(count)
print('max =',max)
```
