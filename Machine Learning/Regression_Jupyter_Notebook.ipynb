```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_copy['gender'] = le.fit_transform(df_copy['gender'])
df_copy['ssc_b'] = le.fit_transform(df_copy['ssc_b'])
df_copy['hsc_b'] = le.fit_transform(df_copy['hsc_b'])
df_copy['workex'] = le.fit_transform(df_copy['workex'])
df_copy['gender'] = le.fit_transform(df_copy['gender'])
df_copy['specialisation'] = le.fit_transform(df_copy['specialisation'])
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.03,random_state=100)
```

```python
# from sklearn.model_selection import train_test_split
# xopt_tr,xopt_te,yopt_tr,yopt_te = train_test_split(x_opt,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x= sc.fit_transform(x)
x_tr= sc.fit_transform(x_tr)
x_te = sc.transform(x_te)
print(x_tr)
print(x_te)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
acc = -10
n = []
for i in range(1,11):
    for j in range(0,150):
        rfr=RandomForestRegressor(n_estimators=i, random_state=j)
        rfr.fit(x_tr,y_tr)
        y_pre = rfr.predict(x_te)
        a = r2_score(y_te,y_pre)
        if(a > acc):
            acc = a
            n.clear()
            n.append(i)
            n.append(j)

rfr=RandomForestRegressor(n_estimators=n[0], random_state=n[1])
rfr.fit(x_tr,y_tr)
y_pre = rfr.predict(x_te)
print(r2_score(y_te,y_pre))
print(mean_absolute_error(y_te,y_pre))
print(mean_squared_error(y_te,y_pre))
print(n)
df = pd.DataFrame({'Actual':y_te, 'Predicted':y_pre})
df
```

```python
y_pre = rfr.predict(x_tr)
print(r2_score(y_tr,y_pre))
print(mean_absolute_error(y_tr,y_pre))
print(mean_squared_error(y_tr,y_pre))
# print(n)
df = pd.DataFrame({'Actual':y_tr, 'Predicted':y_pre})
df
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(15, activation="relu", input_dim=14))
model.add(Dense(15, activation="relu"))
# model.add(Dense(15, activation="relu"))
# model.add(Dense(7, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
model.compile(optimizer="rmsprop", loss='mse', metrics=['mae'])
model.fit(x_tr, y_tr, batch_size=1, epochs=500)
```

```python
y_pe = model.predict(x_te)
print(r2_score(y_te,y_pe))
print(mean_absolute_error(y_te,y_pe))
print(mean_squared_error(y_te,y_pe))
print(y_pe)
print(y_te)
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(x_tr,y_tr)
y_pre = svr.predict(x_te)
print(metrics.r2_score(y_te,y_pre))
df = pd.DataFrame({'Actual':y_te, 'Predicted':y_pre})
df
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics
acc3 = -10
n = []
for i in range(3,118):
    for j in range(1,3):
        knr= KNeighborsRegressor(n_neighbors=i, metric='minkowski', p=j)
        knr.fit(x_tr,y_tr)
        y_p = knr.predict(x_te)
        c = metrics.r2_score(y_te,y_p)
        if(c > acc3):
            acc3 = c
            n.clear()
            n.append(i)
            n.append(j)

knr= KNeighborsRegressor(n_neighbors=n[0], metric='minkowski', p=n[1])
knr.fit(x_tr,y_tr)
y_p = knr.predict(x_te)
print(r2_score(y_te,y_p))
print(mean_absolute_error(y_te,y_p))
print(mean_squared_error(y_te,y_p))
df = pd.DataFrame({'Actual':y_te, 'Predicted':y_p})
df
```

```python
from sklearn.linear_model import LogisticRegression
acc4 = -20
n = []
for i in range(0,101):
    lr = LogisticRegression(random_state=i, solver='saga',max_iter=500, multi_class='auto')
    lr.fit(x_tr,y_tr)
    y_predict = lr.predict(x_te)
    d = metrics.r2_score(y_te,y_predict)
    if(d > acc4):
        acc4 = d
        n.clear()
        n.append(i)

lr = LogisticRegression(random_state=n[0])
lr.fit(x_tr,y_tr)
y_predict = lr.predict(x_te)
print(r2_score(y_te,y_predict))
print(mean_absolute_error(y_te,y_predict))
print(mean_squared_error(y_te,y_predict))
df = pd.DataFrame({'Actual':y_te, 'Predicted':y_predict})
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
lir = LinearRegression()
lir.fit(x_tr,y_tr)
y_pp = lir.predict(x_te)
print(metrics.r2_score(y_te,y_pp))
print(mean_absolute_error(y_te,y_pp))
print(mean_squared_error(y_te,y_pp))
```

```python
from sklearn.tree import DecisionTreeRegressor
acc5 = -10
n = []
for i in range(0,150):
    dtr = DecisionTreeRegressor(random_state=i)
    dtr.fit(x_tr,y_tr)
    y_prr = dtr.predict(x_te)
    x = r2_score(y_te,y_prr)
    if(x > acc5):
        acc5 = x
        n.clear()
        n.append(i)
        
dtr = DecisionTreeRegressor(random_state=n[0])
dtr.fit(x_tr,y_tr)
y_prr = dtr.predict(x_te)
print(r2_score(y_te,y_prr))
print(mean_absolute_error(y_te,y_prr))
print(mean_squared_error(y_te,y_prr))
df = pd.DataFrame({'Actual':y_te, 'Predicted':y_prr})
df
```
